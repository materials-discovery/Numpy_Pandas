# Numpy: 
NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.

```python
import time
import sys

#Difference in the memory occupied by Python in a list and by Numpy using a vector (array)
S = range(1000)
# Python
print(sys.getsizeof(5)*len(S))
# Numpy
D = np.arange(1000)
print(D.size*D.itemsize)
```

**Let's Compare the computation time used by Numpy**

```python
SIZE = 10000000
L1 = range(SIZE)
L2 = range(SIZE)

A1 = np.arange(SIZE)
A2 = np.arange(SIZE)

# Lets measure the time spent by the computer to run this
start = time.time()

result = [(x,y) for x,y in zip(L1,L2)]

print(( time.time()-start)*1000)

start = time.time()

# result = A1 + A2
result = [(x,y) for x,y in zip(L1,L2)]

print((time.time()-start)*1000)

# What about dot product between vectors

# Python lists
arr1 = list(range(1000000))
arr2 = list(range(1000000, 2000000))

# Numpy arrays
arr1_np = np.array(arr1)
arr2_np = np.array(arr2)

#time
start = time.time()
result = 0
for x1, x2 in zip(arr1, arr2):
    result += x1*x2
print(result, (time.time()-start)*10000)    

#time
start = time.time()
result = np.dot(arr1_np, arr2_np)
print(result, (time.time()-start)*10000)

#+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++
# #Example of zip

# numbers = [1, 2, 3]
# letters = ['a', 'b', 'c']
# zipped = zip(numbers, letters)
# print(list(zipped))
```

# **Sum and Multiplication of vectors (arrays) in Python**

4.jpg

```python
a = np.array([(1,2,3), (9,16,81)])
b = np.array([(1,3,3), (1,1,0)])
print(a*b)
print(a-b)
print(a+b)
```

```python
# Elementwise division
print(a/b)
```

```python
# Concatenate two vectors
# Vertical
a = np.array([(1,2,3), (9,16,81)])
b = np.array([(1,2,3), (9,16,81)])
c = np.vstack((a,b))
c
```

```python
# Concatenate two vectors
# Horizontal
a = np.array([(1,2,3), (9,16,81)])
b = np.array([(1,2,3), (9,16,81)])
c = np.hstack((a,b))
c
```

```python
# Using RAVEL to concatenate two vectors
a = np.array([(1,2,3), (4,5,6)])
b = np.ravel(a)
b
```

```python
# Using arange with numpy
import matplotlib.pyplot as plt

np.arange(0, 5, 0.5, dtype=int)

x = np.arange(0,3*np.pi, 0.1)
y = np.sin(x)
# Sine function
plt.plot(x,y)
plt.show()
# Cosine
y = np.cos(x)
plt.plot(x,y)
plt.show()
```
